# Important insights and Visualization

### Insight - 1

How much population purchase which type of subscription.

```python
netflix['subscription_type'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Customer Distribution by monthly plan')
```

### Insight - 2

Mostly Man puchase the premium subscription of netflix.

Mostly Women purchase standard subscription of netflix.

```python
sns.barplot(x=netflix['gender'],y=netflix['watch_hours'],hue=netflix['subscription_type'])
```

### Insight - 3

It shows us that mostly basic subscription people churned.

```python
pd.crosstab(netflix['churned'],netflix['subscription_type'],normalize='columns')*100
```

### Insight - 4

This shows us how much male,female and other are interested in different genre.

```python
pd.crosstab(netflix['gender'],netflix['favorite_genre'],normalize='columns')*100
```

### Insight - 5

in which country,after spending the monthly fee what is thier watch time.Does high pay people watch more or not.

```python
sns.barplot(x=netflix['region'],y=netflix['watch_hours'],hue=netflix['monthly_fee'])
```

In which region what type of subscription people purchase mostly.
