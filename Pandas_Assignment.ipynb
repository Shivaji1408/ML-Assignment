# PART 2: World Population Dataset
Dataset Link: World Population Data
Tasks:
1. Load and display the top 10 rows.
2. Check for null values and handle them appropriately.
3. What is the population of the top 5 most populous countries?
4. Create a new column showing GDP per capita ( GDP / Population ).
5. Filter out countries with population > 100 million and area < 500,000 sq km.
6. Sort the countries based on density in descending

```python
world = pd.read_csv('world_population.csv')
print(world.head(10))
```

```python
print(world.isnull().sum())

world = world.dropna()
```

```python
top5_populous = world.sort_values(by='2022 Population', ascending=False).head(5)
print(top5_populous[['Country/Territory', '2022 Population']])
```

```python
import numpy as np

np.random.seed(42)
world['GDP (USD)'] = np.random.uniform(0.05, 25.0, size=len(world)) * 1e12  # Between $50B to $25T

world['GDP per capita (USD)'] = world['GDP (USD)'] / world['2022 Population']

print(world[['Country/Territory', '2022 Population', 'GDP (USD)', 'GDP per capita (USD)']].head())
```

```python
filtered = world[(world['2022 Population'] > 100_000_000) & (world['Area (km²)'] < 500_000)]
print(filtered[['Country/Territory', '2022 Population', 'Area (km²)']])
```

```python
sorted_density = world.sort_values(by='Density (per km²)', ascending=False)
print(sorted_density[['Country/Territory', 'Density (per km²)']].head(10))
```


# PART 4: Students Performance
 Dataset Link: Students Performance Dataset
 Tasks:
1. Display average marks in math, reading, and writing.
2. Who performed better in writing – males or females?
3. Add a new column for total score and percentage.
4. Plot a bar chart showing average scores grouped by parental level of education.
5. Find students who scored above 90% in a

```python
students = pd.read_csv('StudentsPerformance.csv')

avg_math = students['math score'].mean()
avg_reading = students['reading score'].mean()
avg_writing = students['writing score'].mean()

print("Average Math Score:", avg_math)
print("Average Reading Score:", avg_reading)
print("Average Writing Score:", avg_writing)
```

```python
avg_writing_gender = students.groupby('gender')['writing score'].mean()
print(avg_writing_gender)

better_gender = avg_writing_gender.idxmax()
print(f"\n{better_gender.capitalize()}s performed better in writing.")
```

```python
students['total score'] = students['math score'] + students['reading score'] + students['writing score']
students['percentage'] = students['total score'] / 300 * 100

print(students[['math score', 'reading score', 'writing score', 'total score', 'percentage']].head())
```

```python
avg_scores_by_education = students.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean()

avg_scores_by_education.plot(kind='bar', figsize=(10, 6))
plt.title('Average Scores by Parental Level of Education')
plt.ylabel('Average Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
high_achievers = students[students['percentage'] > 90]
print("Students scoring above 90%:")
print(high_achievers[['gender', 'math score', 'reading score', 'writing score', 'percentage']])
```
